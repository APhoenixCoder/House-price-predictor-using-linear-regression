# **Predicting house prices based on house sizes without using Sci-Kit**

```python
x=[1000, 1500, 2000, 2500, 3000]
y=[50000, 75000, 100000, 125000, 150000]
```

```python
xScaled=[]
for i in range(len(x)):
  xScaled.append(x[i]/1000)
xScaled
```

```python
yScaled=[]
for i in range(len(x)):
  yScaled.append(y[i]/1000)
yScaled
```

```python
slope=50
intercept=0.1
LR=0.000001
epoch=1000
```

```python
for i in range(epoch):
  grad_slope=0
  grad_intercept=0
  for j in range(len(xScaled)):
    predY=slope*xScaled[j]+intercept
    error=predY-yScaled[j]
    grad_slope+= 2 *error* xScaled[j]
    grad_intercept+= 2 *error
  slope -= LR * grad_slope
  intercept -= LR * grad_intercept
  # loss calculation
  if i%100==0:
    loss=0
    for k in range(len(xScaled)):
      pred=slope*xScaled[k]+intercept
      error=pred-yScaled[k]
      # loss=loss+error**2
      loss+=error**2
    print(f"epoch {i} loss {loss}")
```

```python
def predict(x):
  y=slope*(x/1000)+intercept
  return y*1000
```

```python
houseSize=2500
answer=predict(houseSize)
print(answer)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y,label="original data points")
py = [((slope*(X/1000)+intercept)*1000) for X in x]
plt.plot(x,py,color = 'purple',label = "Best Fit Line")
plt.xlabel("houseSizes in square feet")
plt.ylabel("housePrices in SGD")
plt.legend()
plt.show()
```
